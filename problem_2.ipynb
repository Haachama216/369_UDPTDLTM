```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
import numpy as np
import pandas as pd
import os
import re
sns.set(rc={'axes.facecolor':'lightblue', 'figure.facecolor':'lightgreen', 'figure.figsize':(16,9)})
```

### Load and preprocess data

Read all partial data files and merge them into a complete dataframe

```python
def read_all_shop_info(data_dir):
  df_list = []
  filename_list = sorted([os.path.join(data_dir,f) for f in os.listdir(data_dir)])
  for filepath in filename_list:
    df_list.append(pd.read_csv(filepath, sep = '\t'))
  df = pd.concat(df_list)
  df.drop_duplicates(subset='Tên cửa hàng', inplace = True)
  df.reset_index(drop = True, inplace = True)
  return df
```

```python
df = read_all_shop_info('NewData')
df
```

```python
def get_menu_stat(store_name):
  menufile_path = os.path.join(os.getcwd(), 'NewMenuStore', store_name)
  try:
    df = pd.read_csv(menufile_path, encoding = 'utf-8')
    df['Price'] = df['Price'].str.replace(',','').astype(int)
    return df.shape[0], df['Price'].mean()
  except:
    try:
      df = pd.read_csv(menufile_path + '.txt', encoding = 'utf-8')
      df['Price'] =df['Price'].str.replace(',','').astype(int)
      return df.shape[0], df['Price'].mean()
    except:
      print('{} doesnt have menu info file'.format(store_name))
      return 0, 0
```

```python
temp_df = df.apply(lambda x\
  : get_menu_stat(x['Tên cửa hàng']) if x['Cửa hàng có menu'] == 1 else 0, axis = 1, result_type = 'expand')
df['Tổng số lượng sản phẩm'] = temp_df.iloc[:,0].astype(int)
df['Giá trung bình'] = temp_df.iloc[:,1].round(2)
df
```

```python
#Xử lý ngoại lệ số sao
df['Số sao trên Foody'] = df['Số sao trên Foody'].replace('<span>_._</span>','0')
#Xử lý tên cửa hàng
df['Tên cửa hàng'] = df['Tên cửa hàng'].replace('&amp;', 'Và')
#Xử lý lượt đánh giá
df['Số lượt đánh giá trên ShopeeFood'] = df['Số lượt đánh giá trên ShopeeFood'].apply(lambda x: '< 100' if x not in ['100+','500+','999+'] else x)
#Xử lý Phí dịch vụ
df['Phí dịch vụ'] = df['Phí dịch vụ'].apply(lambda x: 'Unknown' if x == 'ShopeeFood</span' else x)
df['Số sao trên Foody'] = df['Số sao trên Foody'].astype('float64')
df['Số lượt đánh giá trên Foody'] = df['Số lượt đánh giá trên Foody'].replace(',','', regex=True).astype(int)
df['Giá thấp nhất'] = df['Giá thấp nhất'].replace(',','', regex=True).astype(int)
df['Giá cao nhất'] = df['Giá cao nhất'].replace(',','', regex=True).astype(int)
df['Thời gian mở cửa'] = df['Thời gian mở cửa'].apply(lambda x: 'Unknown' if x == '0' else x)
df
```

### Phân phối của các shop theo số lượng sản phẩm

Chỉ lấy ra các cửa hàng có menu

```python
df = df[df['Cửa hàng có menu'] == 1]
```

Biểu đồ phân bố số điểm đánh giá trên Foody

```python
g = sns.histplot(data = df, x = 'Số sao trên Foody', binwidth = 1)
g.set_xticks(np.arange(0,11,1))
g.set_title('Biểu đồ phân phối số điểm đánh giá trên Foody')
```

Số lượng các shop được đánh giá cao (điểm Foody > 7.0) theo tổng số lượng sản phẩm

```python
g = sns.histplot(data = df[df['Số sao trên Foody'] > 7.0], x = 'Tổng số lượng sản phẩm')
g.xaxis.set_major_locator(ticker.MultipleLocator(25))
```

Số lượng các shop được đánh giá thấp (điểm trên Foody < 5.0) theo tổng số lượng sản phẩm

```python
g = sns.histplot(data = df[df['Số sao trên Foody'] < 5.0], x = 'Tổng số lượng sản phẩm')
g.xaxis.set_major_locator(ticker.MultipleLocator(25))
```

Để đảm bảo tính đánh giá khách quan hơn, ta sẽ lọc đi các shop có số lượng đánh giá trên ShopeeFood < 100

```python
high_rating_df = df[(df['Số lượt đánh giá trên ShopeeFood'] != '< 100') & (df['Số sao trên Foody'] > 7.0)]
low_rating_df = df[(df['Số lượt đánh giá trên ShopeeFood'] != '< 100') & (df['Số sao trên Foody'] < 5.0)]
```

Số lượng các shop được đánh giá cao (điểm Foody > 7.0) theo tổng số lượng sản phẩm

```python
g = sns.histplot(data = high_rating_df, x = 'Tổng số lượng sản phẩm')
g.xaxis.set_major_locator(ticker.MultipleLocator(25))
```

Số lượng các shop được đánh giá thấp (điểm trên Foody < 5.0) theo tổng số lượng sản phẩm

```python
g = sns.histplot(data = low_rating_df, x = 'Tổng số lượng sản phẩm')
g.xaxis.set_major_locator(ticker.MultipleLocator(25))
```

Số lượng sản phẩm chiếm nhiều nhất ở các shop được đánh giá cao

```python
bins = np.arange(0,high_rating_df['Tổng số lượng sản phẩm'].max(),5)
total_prod_count = high_rating_df['Tổng số lượng sản phẩm'].value_counts(bins = bins)
total_prod_df = total_prod_count.to_frame(name = 'count').reset_index().rename(columns = {'index' : 'total_products'})
total_prod_df['total_products'] = total_prod_df['total_products'].apply(lambda x : '-'.join(re.findall(r'(\d+)\.', str(x))))
total_prod_df
```

```python
g = sns.barplot(data = total_prod_df.head(20), x = 'count', y = 'total_products', orient = 'h', color = 'lightcoral')
g.set_ylabel('Tổng số lượng sản phẩm')
```

Số lượng sản phẩm chiếm nhiều nhất ở các shop bị đánh giá thấp

```python
bins = np.arange(0,low_rating_df['Tổng số lượng sản phẩm'].max(),5)
total_prod_count = low_rating_df['Tổng số lượng sản phẩm'].value_counts(bins = bins)
total_prod_df = total_prod_count.to_frame(name = 'count').reset_index().rename(columns = {'index' : 'total_products'})
total_prod_df['total_products'] = total_prod_df['total_products'].apply(lambda x : '-'.join(re.findall(r'(\d+)\.', str(x))))
total_prod_df
```

```python
g = sns.barplot(data = total_prod_df.head(20), x = 'count', y = 'total_products', orient = 'h', color = 'lightcoral')
g.set_ylabel('Tổng số lượng sản phẩm')
```

### Phân tích trên từng loại cửa hàng

Biểu đồ phân bố của số lượng sản phẩm theo từng cửa hàng

```python
shop_types = ['Quán ăn', 'Café/Dessert', 'Ăn vặt/vỉa hè', 'Nhà hàng', 'Shop Online', 'Tiệm bánh', 'Ăn chay']
new_df = df[['Loại cửa hàng','Tổng số lượng sản phẩm']].reset_index(drop = True)
new_df['Loại cửa hàng'] = new_df['Loại cửa hàng'].apply(lambda x : 'Khác' if x not in shop_types else x)
shop_type_count = new_df['Loại cửa hàng'].value_counts()
plt.pie(shop_type_count.values, labels = shop_type_count.index, autopct = '%0.0f%%', colors = sns.color_palette('pastel'))
```

Boxplot của từng loại shop theo tổng số lượng sản phẩm đối với các shop được đánh giá cao

```python
high_shop_type_df = high_rating_df[['Loại cửa hàng','Tổng số lượng sản phẩm']].reset_index(drop = True)
high_shop_type_df['Loại cửa hàng'] = high_shop_type_df['Loại cửa hàng'].apply(lambda x : 'Khác' if x not in shop_types else x)
orders = high_shop_type_df['Loại cửa hàng'].value_counts().index
g = sns.boxplot(data = high_shop_type_df, x = 'Tổng số lượng sản phẩm', y = 'Loại cửa hàng', order = orders)
g.xaxis.set_major_locator(ticker.MultipleLocator(25))
```

Boxplot của từng loại shop theo tổng số lượng sản phẩm đối với các shop được đánh giá thấp

```python
low_shop_type_df = low_rating_df[['Loại cửa hàng','Tổng số lượng sản phẩm']].reset_index(drop = True)
low_shop_type_df['Loại cửa hàng'] = low_shop_type_df['Loại cửa hàng'].apply(lambda x : 'Khác' if x not in shop_types else x)
orders = low_shop_type_df['Loại cửa hàng'].value_counts().index
g = sns.boxplot(data = low_shop_type_df, x = 'Tổng số lượng sản phẩm', y = 'Loại cửa hàng', order = orders)
g.xaxis.set_major_locator(ticker.MultipleLocator(25))
```

Số lượng các loại shop đối với các shop được đánh giá cao nhưng có số sản phẩm ít (<= 15 sản phẩm)

```python
high_rating_low_prod_df = high_rating_df.loc[high_rating_df['Tổng số lượng sản phẩm'] <= 15, ['Loại cửa hàng', 'Tên cửa hàng', 'Giá trung bình','Tổng số lượng sản phẩm']]
high_rating_low_prod_df.reset_index(drop = True, inplace = True)
high_rating_low_prod_df['Loại cửa hàng'] = high_rating_low_prod_df['Loại cửa hàng'].apply(lambda x : 'Khác' if x not in shop_types else x)
shop_type_count = high_rating_low_prod_df['Loại cửa hàng'].value_counts()
sns.barplot(x = shop_type_count.values, y = shop_type_count.index, orient = 'h')
```

Biểu đồ phân phối giá trung bình của các sản phẩm đối với shop có ít sản phẩm nhưng được đánh giá cao

```python
g = sns.histplot(data = high_rating_low_prod_df, x = 'Giá trung bình')
g.xaxis.set_major_locator(ticker.MultipleLocator(25000))
```

Biểu đồ phân phối giá trung bình của các sản phẩm đối với shop có ít sản phẩm bị đánh giá thấp

```python
low_rating_low_prod_df = low_rating_df.loc[low_rating_df['Tổng số lượng sản phẩm'] <= 15, ['Loại cửa hàng', 'Tên cửa hàng', 'Giá trung bình','Tổng số lượng sản phẩm']]
low_rating_low_prod_df.reset_index(drop = True, inplace = True)
low_rating_low_prod_df['Loại cửa hàng'] = low_rating_low_prod_df['Loại cửa hàng'].apply(lambda x : 'Khác' if x not in shop_types else x)
```

```python
g = sns.histplot(data = low_rating_low_prod_df[low_rating_low_prod_df['Tổng số lượng sản phẩm'] <= 15], x = 'Giá trung bình')
g.xaxis.set_major_locator(ticker.MultipleLocator(25000))
```
